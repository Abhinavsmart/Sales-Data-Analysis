### importing important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
from glob import glob
```

### Merging all months of the data

```python
df= pd.read_csv("C:/Users/Abhinav kumar/Sales_April_2019.csv")
df
```

```python
all_files = glob('Sales*') 
all_files
```

```python
df= pd.concat((pd.read_csv(file) for file in all_files), ignore_index= True) 
```

```python
df.sort_values("Order Date")
```

```python
df.info()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df=df.sort_values("Order Date").reset_index(drop= True)
df
```

### Find Or and delete it.

```python
temp_df= df[df['Order Date'].str[0:2]=="Or"]
```

```python
temp_df
```

```python
df= df[df['Order Date'].str[0:2]!="Or"]
df.head()
```

### Drop rows of NaN value

```python
nan_df= df[df.isna().any(axis=1)]
nan_df
```

```python
df=df.dropna(how='all')
df.head()
```

### Redefine data type of column in appropriate way 

```python
df['Quantity Ordered']= pd.to_numeric(df['Quantity Ordered'])
df['Price Each']= pd.to_numeric(df['Price Each'])
```

### Add new column Month

```python
df['Month']= df['Order Date'].str[:2]
df['Month']= df['Month'].astype('int64')
```

```python
df.head()
```

### Add a sales Column

```python
df['Sales']= df['Quantity Ordered']*df['Price Each']
df.head(10)
```

### What was the best month for sales? How much was earned that month?

```python
Sales_Data=df.groupby('Month').sum()
Sales_Data
```

```python
max_sales=df.groupby('Month').sum()['Sales'].max()
max_sales
```

```python
months= range(1,13)
plt.bar(months,Sales_Data['Sales'] )
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Months')
plt.title("Sales Per Month")
plt.show()
```

### Add a City Column

```python
df['City']=df['Purchase Address'].apply(lambda x: x.split(',')[1])
df.head()
```

### What city had the highest no of Sales

```python
Max_city=df.groupby('City').sum()
Max_city
```

```python
Max_city.idxmax()
```

```python
cities= [city for city , df in df.groupby('City')]
plt.bar(cities, Max_city['Sales'])
plt.xticks('cities', rotation='vertical', size=8)
plt.xlabel('City')
plt.ylabel('Sales in USD')
plt.title("Sales in City")
plt.show()

#why city xticks is not showing city name
```

### what time shoud we display advertisement to maximize liklihood of customer's buying products

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df.head()
```

```python
df['Hour']=df['Order Date'].dt.hour
df['Minute']=df['Order Date'].dt.minute
df.head()
```

```python
hours= [hour for hour, df in df.groupby('Hour')]
plt.plot(hours,df.groupby(['Hour']).count())
plt.xlabel('Hour')
plt.ylabel('No of orders')
plt.xticks(hours)
plt.grid()
```

### What products are most often sold together ?

```python
df=df[df['Order ID'].duplicated(keep= False)]
df['Combined']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df_n=df[['Order ID', 'Combined']].drop_duplicates()
df_n   
```

```python
from itertools import combinations
from collections import Counter
```

```python
count= Counter()
for row in df_n['Combined']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):
    print(key, value)
```
